```python
from google_play_scraper import app
import pandas as pd
import numpy as np
```

```python
packagename='pl.lukok.chess'

from google_play_scraper import Sort, reviews


us_reviews, continuation_token = reviews(
    packagename,
    count = 20000,
)


df_busu = pd.DataFrame(np.array(us_reviews),columns=['review'])# or
df_busu = df_busu.join(pd.DataFrame(df_busu.pop('review').tolist()))
df_busu
```

```python
file=pd.read_csv('/Users/LENOVO/Project/'+packagename+'.csv')
file
```

```python
count=0
for index,row in df_busu.iterrows():
    file.at[count,"Package name"]=packagename
    file.at[count,"Reviewer name"]=row["userName"]
    file.at[count,"Review"]=row["content"]
    file.at[count,"Rating"]=row["score"]
    count=count+1
file
```

```python
file.to_csv('/Users/LENOVO/Project/'+packagename+'.csv',index=False)
file
```
