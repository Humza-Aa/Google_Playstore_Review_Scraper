```python
#Task 2
import sys
!{sys.executable} -m pip install nltk
import nltk
nltk.download()
import pandas as pd
import string
import numpy as np
import re
from num2words import num2words
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
from nltk.stem import WordNetLemmatizer
```

```python
#Task 2
AllApps = pd.read_csv('/Users/LENOVO/Project/MyAppDescriptions.csv', skipinitialspace = True)
packages = AllApps.filter(['Package Name'])
for (indexe, row) in packages.iterrows():
    PackageName = packages["Package Name"][indexe]
    app = pd.read_csv('/Users/LENOVO/Project/'+PackageName+'.csv', chunksize = 3000)
    app = pd.concat(app)
    #Reviews = app.filter(['Review'])
    lemmatizer = WordNetLemmatizer()
    print('')
    print(f'Package Name: {PackageName}')
    print('')
    
    def ignorenan(string2):   
        if string2 is np.nan :
            return ''
        else:
            return string2
    def textpreprocessing(review):
        i = 0
        stringsplit = review.split()
        for word in stringsplit:
            if word.isnumeric():
                word = num2words(word)
                stringsplit[i] = word
            if stringsplit[i] not in stop_words:
                #print(word)
                word = lemmatizer.lemmatize(word)
            i = i + 1
        return " ".join(stringsplit)
    
    app['Review'] = app['Review'].apply(ignorenan)
    app['Review'] = app['Review'].apply(lambda review: review.lower())
    #app['Review'] = app['Review'].apply(lambda review: " ".join([num2words(word) for word in review.split() if word.isnumeric()]))
    app['Review'] = app['Review'].apply(lambda review: "".join([char for char in review if char not in string.punctuation]))
    app['Review'] = app['Review'].apply(lambda review: re.sub(r"[^a-zA-Z0-9 ]", "", review))
    app['Review'] = app['Review'].apply(lambda review: " ".join([word for word in review.split() if word not in stop_words]))
    #app['Review'] = app['Review'].apply(lambda review: " ".join([lemmatizer.lemmatize(word) for word in review.split()]))
    app['Review'] = app['Review'].apply(textpreprocessing)
    #print(app['Review'])
    app.to_csv(f'{PackageName}.csv', index = False)

    numReviews = 0
    while numReviews <= 15:
        print(f'{numReviews}: {app["Review"][numReviews]}')
        numReviews = 1 + numReviews
```

```python
#Task 2 bonus:

# The choice of stopwords does have an impact on the analysis of the reviews as we want to eliminate only the common words and
# analize the most important parts of a given review 
# The risk of using a non-customized list of stop words in analyzing mobile app reviews for software requirements elicitation
# is that the default stopword list might not have the unique stopwords that a user might express, such as "YEAAAAAAA" or "COOOL"
```

```python
#Task 3 III
import sys
!{sys.executable} -m pip install vaderSentiment
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()

AllApps = pd.read_csv('/Users/LENOVO/Project/MyAppDescriptions.csv', skipinitialspace = True)
#packages = AllApps.filter(['Package Name'])
pn = []
rv = []
Rat = []

def GetPolarity(PackageName):
    global pn, rv, Rat
    #PackageName = #packages["Package Name"][indexe]
    app = pd.read_csv('/Users/LENOVO/Project/'+PackageName+'.csv', chunksize = 3000)
    app = pd.concat(app)
    #print(app2)
    #Reviews = app.filter(['Review'])
    def ignorenan(string2):   
            if string2 is np.nan :
                return ''
            else:
                return string2

    app['Review'] = app['Review'].apply(ignorenan)
    app['Rating']= app['Review'].apply(lambda ProcessedReview: (analyzer.polarity_scores(ProcessedReview))["compound"])
    app = app.rename(columns={'Rating': 'Polarity'})
    app.drop('Reviewer name', inplace=True, axis=1)
    pn = np.concatenate((pn,  app['Package name']))
    rv = np.concatenate((rv,  app['Review']))
    Rat = np.concatenate((Rat,  app['Polarity']))

    
AllApps['Package Name'].apply(GetPolarity)

df = {'Package Name': pn, 'Reveiw': rv, 'Polarity': Rat}
df = pd.DataFrame(df)
print(df)
```

```python
#Task 4 Reference:https://rstudio-pubs-static.s3.amazonaws.com/79360_850b2a69980c4488b1db95987a24867a.html#applying-the-lda-model
import sys
!conda install --yes --prefix {sys.prefix} gensim
from gensim import corpora, models
from nltk.tokenize import RegexpTokenizer
tokenizer = RegexpTokenizer(r'\w+')
```

```python
#Task 4 Cont.
dictionary2 = ''
PackageName = 'com.chess.clock'
dictionary = ''

app = pd.read_csv('/Users/LENOVO/Project/'+PackageName+'.csv', chunksize = 3000)
app = pd.concat(app)

def ignorenan(string2):   
        if string2 is np.nan :
            return ''
        else:
            return string2

def concat(string): # sting = array of strings
    global dictionary2
    dictionary2 = dictionary2 + ' ' + string
    return dictionary2


app['Review'] = app['Review'].apply(ignorenan)
dictionary = app['Review'].apply(concat)
dictionary2 = tokenizer.tokenize(dictionary2) 
print(dictionary2)
```

```python
from gensim import corpora, models
d = corpora.Dictionary([dictionary2])
corpus = [d.doc2bow([word]) for word in dictionary2]
ldamodel = models.ldamodel.LdaModel(corpus, num_topics=10, id2word = d, passes=20)
```

```python
print(ldamodel.print_topics(num_topics=10, num_words=7))
```

```python
topics = ldamodel.print_topics(num_topics=10, num_words=7)
pd.options.display.max_colwidth=300

df = {'Topics': topics}
#pd.options.display.max_colwidth=15
df = pd.DataFrame(df)
print(df)
```

```python
#Task 4 IV
AllApps = pd.read_csv('/Users/LENOVO/Project/MyAppDescriptions.csv', skipinitialspace = True)

#for (indexe, row) in packages.iterrows():
def GetTopics(PackageName):

    di2 = ''
    #packageName = 'com.chess'
    di = ''
    
    app = pd.read_csv('/Users/LENOVO/Project/'+PackageName+'.csv', chunksize = 3000)
    app = pd.concat(app)

    def ignorenan(string2):   
            if string2 is np.nan :
                return ' '
            else:
                return string2

    def concat(string): # sting = array of strings
        nonlocal di2
        #print(type(dictionary2))
        di2 = di2 + ' ' + string
        #return di2


    app['Review'] = app['Review'].apply(ignorenan)
    #print(dictionary2)
    #print(app['Review'])
    di = app['Review'].apply(concat)
    di2 = tokenizer.tokenize(di2) 
    d = corpora.Dictionary([di2])
    corpus = [d.doc2bow([word]) for word in di2]
    ldamodel = models.ldamodel.LdaModel(corpus, num_topics=5, id2word = d, passes=20)
    topics = ldamodel.print_topics(num_topics=10, num_words=7)
    df = {'Topics': topics}
    df = pd.DataFrame(df)
    print(df)    
    
AllApps['Package Name'].apply(GetTopics)
#print(di2)
#print(dictionary)
```
